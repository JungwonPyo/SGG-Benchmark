## Extract object classes

```python
# If COCO-format
import json

source_file = "../datasets/PSG/coco_format/val/_annotations.coco.json"

data = json.load(open(source_file))
classes = data["categories"]
classes = [c["name"] for c in classes]
classes = sorted(classes)
print(classes)
```

```python
# If original .h5 format
import json

source_file = "../datasets/vg/VG-SGG-dicts-with-attri.json"
# Load the custom classes
with open(source_file) as f:
    data = json.load(f)
classes = list(data['idx_to_label'].values())
print(classes)
```

```python
# YOLOE
from ultralytics import YOLO, YOLOE

# Initialize a YOLOE model
model = YOLO("yoloe-11l.yaml").load("yoloe-11l-seg.pt")  # or select yoloe-11s/m-seg.pt for different sizes

model.eval()
model.ckpt = {}

print("YoloE will be setup for the following classes: ", classes)

model.set_classes(classes, model.get_text_pe(classes))

model.save("custom_yoloe11l.pt")
```

```python
# Taken from https://docs.ultralytics.com/models/yolo-world/
# This script is used to save the model with the defined offline vocabulary

from ultralytics import YOLOWorld
import json

model = YOLOWorld("yolov8x-worldv2.pt")  # or select yolov8m/l-world.pt or another variant

print("Yolo-World will be setup for the following classes: ", classes)
# Define custom classes
model.set_classes(classes)

# Save the model with the defined offline vocabulary
model.save("custom_yolov8x_vg150.pt")
```
